```python
df = pd.read_csv("Base - Desligamentos.csv", delimiter="|",  header=1)
```

# Tratamento de dados faltantes

```python
col_nulls = df.isnull().sum()[df.isnull().sum()>0].keys().tolist()
df.isnull().sum()[df.isnull().sum()>0]
```

```python
df[col_nulls].select_dtypes(include=['number']).head()
```

```python
imputer = KNNImputer(n_neighbors=2)
df[col_nulls] = pd.DataFrame(imputer.fit_transform(df[col_nulls]), columns = col_nulls)
df.isnull().sum()[df.isnull().sum()>0]
```

# Padronização das colunas

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ',"_")
```

```python
df.head()
```

# Padronização do valor TARGET

```python
df.desligado.value_counts()
```

```python
df.desligado = df.desligado.str.lower()
df.desligado = df.desligado.str.strip()
df['target'] = df.desligado.replace({"no": 0, "yes": 1})
df.target.value_counts()
```

```python
df.info()
```

# Análise do TARGET

```python
data = df.desligado.value_counts()

pie, ax = plt.subplots(figsize=[10,6])
labels = data.keys()
plt.pie(x=data, autopct="%.1f%%", explode=[0.05]*2, labels=labels, pctdistance=0.5)
plt.title("Distribuição de desligado da empresa", fontsize=14)
```

**Apenas 16.1% se desligaram da empresa**

# Entendimento dos dados

```python
plt.figure(figsize=(20,7))
plt.title("Distribuição de desligados por campo de educação",fontsize=40)
sns.countplot(data=df, x="campo_da_educacao", hue="desligado")
```

```python
df_ = df.groupby(['campo_da_educacao','desligado']).agg({'idade': 'count'}).groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).reset_index()
# Draw a nested barplot by species and sex
plot_order = df_.sort_values(by=['idade'], ascending=False).idade.values
g = sns.catplot(
    data=df_.sort_values('idade'), kind="bar",
    x="campo_da_educacao", y="idade", hue="desligado",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Percentual (%) de diferença")
g.legend.set_title("")

g.fig.set_size_inches(20, 10)
g.add_legend(frameon=True)
g.fig.suptitle('Percentual (%) de desligados por campo de educação', fontsize=25)
```

# Conclusões :
***Grupo com maior percentual de desligamentos da empresa é do de RH***
