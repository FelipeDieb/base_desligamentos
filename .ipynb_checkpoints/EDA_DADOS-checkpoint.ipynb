# Tratamento de dados faltantes

```python
col_nulls = df.isnull().sum()[df.isnull().sum()>0].keys().tolist()
df.isnull().sum()[df.isnull().sum()>0]
```

```python
def missing_data(data):
    # Count number of missing value in a column
    total = data.isnull().sum()           
    
    # Get Percentage of missing values
    percent = (data.isnull().sum()/data.isnull().count()*100)   
    temp = pd.concat([total, percent], axis=1, keys=['Total', 'Percent(%)'])

    # Create a Type column, that indicates the data-type of the column.
    types = []
    for col in data.columns:
        dtype = str(data[col].dtype)
        types.append(dtype)
    temp['Types'] = types

    return(np.transpose(temp))

missing_data(df[col_nulls])
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df[col_nulls].isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
df.dropna(axis=1, thresh = int(0.8*df.shape[0]), inplace=True)
```

```python
col_nulls = df.isnull().sum()[df.isnull().sum()>0].keys().tolist()
missing_data(df[col_nulls])
```

# Removendo outliers

```python
def get_iqr_values(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25) # 25th percentile / 1st quartile
    q3 = df_in[col_name].quantile(0.75) # 7th percentile / 3rd quartile
    iqr = q3-q1 #Interquartile range
    minimum = q1-1.5*iqr # The minimum value or the |- marker in the box plot
    maximum = q3+1.5*iqr # The maximum value or the -| marker in the box plot
    return minimum, maximum

def remove_outliers(df_in, col_name):
    minimum, maximum = get_iqr_values(df_in, col_name)
    print("Valor Minimo {}, Valor Máximo {}".format(minimum, maximum))
    df_out = df_in.loc[(df_in[col_name] > minimum) & (df_in[col_name] < maximum)]
    return df_out
```

```python
df = remove_outliers(df,'idade')
df = remove_outliers(df,'renda_mensal')
```

<center class="font-size:50px"><h1 >Caracteristicas gerais por desligamento</h1></center>

```python

l = ['satisfacao_com_relacionamento','satisfacao_com_trabalho','satisfacao_com_ambiente',
     'anos_com_atual_gestor','anos_na_empresa','envolvimento_com_o_trabalho',
     'equilibrio_vida_trabalho','nivel_educacional','desempenho','treinamentos_no_ultimo_ano']
data = df.groupby('desligado')[l].agg(pd.Series.mode)
categories = data.keys()

fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=data.loc['no'].values.tolist(),
      theta=categories,
      fill='toself',
      name='Não desligados'
))
fig.add_trace(go.Scatterpolar(
      r=data.loc['yes'].values.tolist(),
      theta=categories,
      fill='toself',
      name='Desligados'
))

fig.update_layout(
    title="Características dos grupos de desligados e não desligados",
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, max(data.loc['no'].values.tolist())]
    )),
  showlegend=True
)
fig.show(renderer="svg", width=950, height=500)
```

- **É visto que as caracteristicas que mais se sobresaem no grupo de não desligados são: anos na empresa, anos com o atual gestor, satisfacao com ambiente e satisfacao com trabalho**
- **Pessoas com pouco tempo de empresa e tempo com o atual gestor terá que ter cuidados atenção especial pela gestão**

<center class="font-size:50px"><h1 >Desempenho vs tempo de empresa</h1></center>

```python
df_ = df.groupby(['desempenho','desligado']).agg({'idade': 'count'}).groupby(level=0).apply(lambda x: 100 * x / float(x.sum())).reset_index()
# Draw a nested barplot by species and sex
plot_order = df_.sort_values(by=['idade'], ascending=False).idade.values
g = sns.catplot(
    data=df_.sort_values('idade'), kind="bar",
    x="desempenho", y="idade", hue="desligado",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Percentual (%) de diferença")
g.legend.set_title("")

g.fig.set_size_inches(20, 7)
g.add_legend(frameon=True)
g.fig.suptitle('Percentual (%) de desligados por desempenho', fontsize=25)
```

```python

g = sns.FacetGrid(df, col="desempenho", hue="desligado")
g.map(sns.histplot, "anos_na_empresa")
g.fig.set_size_inches(14, 7)
g.add_legend(frameon=True)
g.set_yticklabels(size = 50)
g.set_xticklabels( size = 12)
g.fig.suptitle('Distribuição da anos de empresa por desempenho e desligados', fontsize=25)
```

- **É visto que o grupo de desempenho 2, não duram muito tempo na empresa, a sua grande maioria é desligada**
- **No desempenho 3, um funcionário com seus primeiros anos costumam se desligar mais**
- **É notório que os primeiros anos da empresa existem uma tendência maior de desligamento**
- **É notório que o grupo de desempenho 2, tem um potencial muito elevado de desligados, ~80% deste grupos estão desligado da empresa**
